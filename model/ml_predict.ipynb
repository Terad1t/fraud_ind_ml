```python
import joblib
import pandas as pd

model = joblib.load("model/fraud_model.pkl")

new_transaction = pd.DataFrame([{
    "valor": 4500,
    "hora": 2,
    "pais": "US",
    "dispositivo": "mobile",
    "tentativas": 3,
    "cartao_presente": True
}])

pred = model.predict(new_transaction)[0]
prob = model.predict_proba(new_transaction)[0][1]

print("resultado: ")

if pred == 1:
    print(f"FRAUDE detectada! (probabilidade: {prob:.2f})")
else:
    print(f"Transação normal (probabilidade de fraude: {prob:.2f})")
```
